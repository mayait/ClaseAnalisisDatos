```python
import lyricsgenius

def download_song_lyrics(api_key, artist_name, song_title):
    # Initialize the Genius object with the API key
    genius = lyricsgenius.Genius(api_key)

    # Search for the song
    song = genius.search_song(song_title, artist_name)

    # Set the timeout duration (in seconds)
    genius.timeout = 15

    if song:
        return song.lyrics
    else:
        return None

# Example usage:
ARTIST_NAME = 'metallica'
SONG_TITLE = 'one'

lyrics = download_song_lyrics(API_KEY, ARTIST_NAME, SONG_TITLE)
print(lyrics)
```

```python
metallica_songs = [
    "Enter Sandman",
    "Master of Puppets",
    "One",
    "Fade to Black",
    "The Unforgiven",
    "For Whom the Bell Tolls",
    "Nothing Else Matters",
    "Seek & Destroy",
    "Creeping Death",
    "Battery",
    "Ride the Lightning",
    "Welcome Home (Sanitarium)",
    "Blackened",
    "Sad But True",
    "Whiplash",
    "Orion",
    "Harvester of Sorrow",
    "The Four Horsemen",
    "Fuel",
    "The Day That Never Comes"
]

ARTIST_NAME = 'Metallica'
```

```python
# Create an empty DataFrame with the columns "song_title", "artist_name", and "lyrics"
lyrics_df = pd.DataFrame(columns=["song_title", "artist_name", "lyrics"])

for song_title in metallica_songs:
    print(f"Downloading lyrics for '{song_title}'...")
    lyrics = download_song_lyrics(API_KEY, ARTIST_NAME, song_title)
    
    if lyrics:
        print(f"Lyrics for '{song_title}' downloaded.")
        # Append the new row to the DataFrame
        lyrics_df = lyrics_df.append({"song_title": song_title, "artist_name": ARTIST_NAME, "lyrics": lyrics}, ignore_index=True)
    else:
        print(f"Lyrics for '{song_title}' not found.\n")

# Print the DataFrame
print(lyrics_df)
```
